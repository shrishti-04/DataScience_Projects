```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_title = pd.read_csv('Movie_Id_Titles')
movie_title.head()
```

```python
movie_df = pd.merge(df, movie_title, on='item_id')
movie_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
%matplotlib inline
```

```python
movie_df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
movie_df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
titles = pd.DataFrame(movie_df.groupby('title')['rating'].mean())
titles.head()
```

```python
titles['no_of_ratings'] = pd.DataFrame(movie_df.groupby('title')['rating'].count())
titles.head()
```

```python
titles['no_of_ratings'].hist(bins=70)
```

```python
titles['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='no_of_ratings', data=titles, alpha=0.5)
```

```python
moviemat = movie_df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat.head()
```

```python
titles.sort_values('no_of_ratings', ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
toystory_user_ratings = moviemat['Toy Story (1995)']
```

```python
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars.head()
```

```python
similar_to_toystory = moviemat.corrwith(toystory_user_ratings)
similar_to_toystory
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlations'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlations', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(titles['no_of_ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['no_of_ratings'] > 100].sort_values('Correlations', ascending=False).head()
```

```python
corr_toystory = pd.DataFrame(similar_to_toystory, columns=['Correlations'])
corr_toystory.dropna(inplace=True)
corr_toystory.head()
```

```python
corr_toystory = corr_toystory.join(titles['no_of_ratings'])
corr_toystory.head()
```

```python
corr_toystory[corr_toystory['no_of_ratings'] > 100].sort_values('Correlations', ascending=False).head()
```
